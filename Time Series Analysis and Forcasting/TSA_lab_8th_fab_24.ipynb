```python
df = pd.read_csv(r"C:\Users\raval\jupyter_notebook\time-series-analysis-and-forcasting\Monthly Steam Usage and Average Ambient Temperature.csv")
```

```python
# Extract features (temperature) and target variable (usage)
X = np.column_stack((np.ones(len(df)), df['Temperature']))
y = df['Usage'].values

# Calculate coefficients using the normal equation
theta = np.linalg.inv(X.T @ X) @ X.T @ y

# Print the coefficients
intercept, slope = theta
print(f'Intercept (theta_0): {intercept:.4f}')
print(f'Slope (theta_1): {slope:.4f}')
```

```python
plt.scatter(df['Temperature'], df['Usage'], label='Data Points')
regression_line = X @ theta
plt.plot(df['Temperature'], regression_line, color='red', label='Regression Line')

# Add labels and title
plt.xlabel('Temperature')
plt.ylabel('Usage')
plt.title('Linear Regression')
plt.legend()

# Show the plot
plt.show()
```

```python
# Assign numerical values to months
month_values = np.arange(len(df['Month']))

# Extract features (temperature) and target variable (usage)
X = np.column_stack((np.ones(len(df)), df['Temperature']))
y = df['Usage'].values

# Calculate coefficients using the normal equation
theta = np.linalg.inv(X.T @ X) @ X.T @ y

# Plot the 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(month_values, df['Usage'], df['Temperature'], c='blue', marker='o', label='Data Points')

# Plot the regression line
regression_line_values = X @ theta
ax.plot(month_values, regression_line_values, df['Temperature'], color='red', label='Regression Line')

# Add labels and title
ax.set_xlabel('Month')
ax.set_ylabel('Usage')
ax.set_zlabel('Temperature')
ax.set_title('3D Linear Regression with Time, Usage, and Temperature')

# Customize x-axis ticks to show months
ax.set_xticks(month_values)
ax.set_xticklabels(df['Month'], rotation=45, ha='right')

# Show the plot
plt.legend()
plt.show()
```

```python
# Plot the scatter plot
plt.scatter(df['Temperature'], df['Usage'], c='blue', marker='o', label='Data Points')

# Plot the regression line
regression_line = X @ theta
plt.plot(df['Temperature'], regression_line, color='red', label='Regression Line')

# Annotate each point with the corresponding month
for i, month in enumerate(df['Month']):
    plt.annotate(month, (df['Temperature'][i], df['Usage'][i]), textcoords="offset points", xytext=(0, 5), ha='center')

# Add labels and title
plt.xlabel('Temperature')
plt.ylabel('Usage')
plt.title('Linear Regression with Data Points and Regression Line')

# Show the plot
plt.legend()
plt.show()
```
