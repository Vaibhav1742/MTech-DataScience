```python
import numpy as np
import pandas as pd

# Given data
amounts = [9, 8, 9, 12, 9, 12, 11, 7, 13, 9, 11, 10]

# Calculate mean of the data
mean_value = np.mean(amounts)

# Calculate differences between each value and the mean
differences = [value - mean_value for value in amounts]

# Calculate square of differences
squared_differences = [diff ** 2 for diff in differences]

# Create DataFrame
df = pd.DataFrame({'Supplier': range(1, len(amounts)+1),
                   'Amount': amounts,
                   'Difference': differences,
                   'Squared Difference': squared_differences})

# Display the DataFrame
df
```

```python
df["mean_value_12_diff"] =  [value - 12 for value in amounts]
df["squered_mean_value_12_diff"] = [diff ** 2 for diff in df["mean_value_12_diff"]]
```

```python
def moving_average(data, window_size):

  moving_averages = []
  for i in range(window_size - 1, len(data)):
    window = data[i - window_size + 1: i + 1]
    average = sum(window) / window_size
    moving_averages.append(average)

  return moving_averages
```

```python
moving_average(df["Amount"],3)
```

```python
def moving_average_with_nan(data, window_size):
    moving_averages = [None] * window_size  # Initialize with NaN values
    for i in range(window_size, len(data) + 1):
        window = data[i - window_size:i]
        average = sum(window) / window_size
        moving_averages.append(average)
    return moving_averages

# # Assuming your data is in a list format
# data = [8, 9, 10, 11, 10, 9, 11, 7, 13, 9, 11, 10]

# window_size = 3
# result = moving_average_with_nan(data, window_size)

# print(result)
```

```python
# df["Moving_Avg_3_scratch"] = df["Amount"].apply(lambda x: moving_average_with_nan(df["Amount"].iloc[max(0, x-window_size+1):x+1], window_size))

# Apply the moving_average_with_nan function on the "Amount" column
df["Moving_Avg_3_scratch"] = df.index.to_series().apply(lambda x: moving_average_with_nan(df["Amount"].values[:x], window_size))

# Flatten the list of lists and convert to DataFrame column
df["Moving_Avg_3_scratch"] = df["Moving_Avg_3_scratch"].apply(lambda x: x[-1])

print(df["Moving_Avg_3_scratch"])
```

```python
# Apply the moving_average_with_nan function on the "Amount" column
df["Moving_Avg_3_scratch_centered"] = df.index.to_series().apply(lambda x: moving_average_with_nan(df["Amount"].values[:x+2], window_size))

# Flatten the list of lists and convert to DataFrame column
df["Moving_Avg_3_scratch_centered"] = df["Moving_Avg_3_scratch_centered"].apply(lambda x: x[-1])

print(df["Moving_Avg_3_scratch_centered"])
```

```python
# Create a new column to store the calculated medians
df['new_column'] = None

# Calculate medians in groups of three
for i in range(2, len(df)):
    median_value = df['Amount'].iloc[i-2:i+1].median()
    df.at[i+1, 'new_column'] = median_value

# Display the DataFrame
print(df["new_column"])
```

```python
df['new_column1'] = None

# Calculate medians in groups of three
for i in range(2, len(df)):
    median_value1 = df['Amount'].iloc[i-2:i+1].mean()
    df.at[i+1, 'new_column1'] = median_value1

# Display the DataFrame
print(df["new_column1"])
```

```python
df['new_column2'] = None

# Calculate medians in groups of three
for i in range(2, len(df)):
    median_value2 = df['Amount'].iloc[i-2:i+1].mean()
    df.at[i-1, 'new_column2'] = median_value2

# Display the DataFrame
print(df["new_column2"])
```

```python
hanning_window = np.hanning(window_size)

# Create a new column to store the filtered values
df['filtered_column'] = None

# Apply Hanning filter
for i in range(1, len(df)-1):
    # Apply Hanning filter to each value and store it in the new column
    df.at[i, 'filtered_column'] = np.sum(df['Amount'].iloc[i-1:i+2] * hanning_window)

# Display the DataFrame
print(df['filtered_column'])
```
