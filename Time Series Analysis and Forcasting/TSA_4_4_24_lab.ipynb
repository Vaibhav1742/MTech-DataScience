```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf

# Plot the autocorrelation function
plot_acf(df['Southern Osc'], lags=100)  # Adjust lags as needed

plot_pacf(df['Southern Osc'], lags=100)  # Adjust lags as needed

plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('Autocorrelation Function of Southern Oscillation Data')
plt.show()
```

```python
# Plot both ACF and PACF
fig, axes = plt.subplots(2, 1, figsize=(10, 8))

# Plot ACF
plot_acf(df['Southern Osc'], lags=100, ax=axes[0])
axes[0].set_title('Autocorrelation Function')

# Plot PACF
plot_acf(df['Southern Osc'], lags=100, ax=axes[1])
axes[1].set_title('Partial Autocorrelation Function')

plt.tight_layout()
plt.show()
```

```python
def autocorr(x):
    n = len(x)
    mean = np.mean(x)
    corr = np.correlate(x - mean, x - mean, mode='full')
    corr /= np.max(corr)
    return corr[n - 1:]


# Compute the autocorrelation function
corr = autocorr(df['Southern Osc'])

# Plot the autocorrelation function
plt.plot(corr)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.show()
```
